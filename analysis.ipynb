```python
import pandas as pd
import matplotlib.pyplot

DB_path = 'data.csv'

gamedata = pd.read_csv(DB_path)
```

```python
gamedata.head()
```

```python
time_vs_range_correlation = gamedata.groupby('RunTime')['PlaneRange'].mean()
graph_title = 'Time vs Range Correlation'
time_vs_range_correlation.plot(kind='line', title=graph_title, ylabel='Range', xlabel='Time')
```

The graph above represents the displacement-time relationship of the average statistics of a player playing highflyer. It is clear that as the more time is elapsed on a run, there is a positive proportional linear relationship between the distance flown and the time elapsed
